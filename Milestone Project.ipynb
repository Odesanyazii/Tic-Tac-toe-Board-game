```python
from IPython.display import clear_output



#I want to check if the user wants to keep playing after someone has won or a draw so i just initialize this for now
game_on_choice = "Y"


#Now while the users want to keep playing, i want the game to run so I'll just make the game run under a while loop. 
#If the game on choice changes to No at the end of the game, the game will stop 

while game_on_choice == "Y":
    
    #Here i want the page to be empty everytime the game starts so i'll clear previous output
    clear_output(wait=True)
    
    #first i show the gamers the tictactoe table positions
    
    print("The tic-tac-toe table looks like this. Please take note of the positions! Enjoy!! \n")

    sample_display_board_row1 = " 1 | 2 | 3 |\n"
    sample_display_board_row2 = " 4 | 4 | 6 |\n"
    sample_display_board_row3 = " 7 | 8 | 9 |\n"
        
        
        
    print(sample_display_board_row1)
    print(sample_display_board_row2)
    print(sample_display_board_row3)
    
    #My approach is to assign the positions of the tictactoe table to variables a - i
    #Then i initialize the variables to be empty spaces so the tictactoe table looks empty

    a = " "
    b = " "
    c = " "
    d = " "
    e = " "
    f = " "
    g = " "
    h = " "
    i = " "

    #Next i make an empty list. My intention is to later save positions on the board that have been marked to this list
    #This is because I want to assign my variables to one of this positions later on

    initial_positions= list(" " * 9)
    
    #Okay, now i need to know if Player 1 is X or O so i can mark the positions appropriately
    #So i make a function that asks the players and returns their answer
    def mark_choice(): 
            choice = "F"
            while choice not in ["X", "O"]:
                choice = input("For Player 1 are you X or O?  ")
                if choice not in ["X", "O"]:
                    print ("Please try again! \n")
            return choice

    first_player_mark = mark_choice()


    #Here because i know what mark player 1 has chosen, I need to keep a play count track so i know whose turn it is
    play_count=0
    
    #Now i just make the actual board they'll be playing on and print it for them to see
    
    display_board_row1 = " {a} | {b} | {c} |\n"
    display_board_row2 = " {d} | {e} | {f} |\n"
    display_board_row3 = " {g} | {h} | {i} |\n"
        
        
        
    print(display_board_row1.format(a=a, b=b, c=c))
    print(display_board_row2.format(d=d, e=e, f=f))
    print(display_board_row3.format(g=g, h=h, i=i))

    
    #I want to check after each entry if someone has won so i just initialize this for now
   
    game_win_check = "N"
    
    #Okay I want the game to keep running until all the positions have been filled or until someone wins right? 
    #so i just make a while loop to keep the game running until no position is empty or someone wins
    
    while " " in [a, b, c, d, e, f, g, h, i] and game_win_check == "N":     

    
        #Okay now I make a function that keeps updating the play count so I know whose turn it is
        
        def play_counter(play_count):
            play_count = play_count + 1
            return play_count

        play_count = play_counter(play_count)
        
        #Then depending on whose turn it is, I make the mark either X or O

        def mark_type(play_count, first_player_mark): 
            if first_player_mark == "X": 
                if play_count %2 == 0: 
                    choice = "O"
                else: 
                    choice = "X"
            
            if first_player_mark == "O": 
                if play_count %2 == 0: 
                    choice = "X"
                else: 
                    choice = "O"
            
            return choice

        mark = mark_type(play_count, first_player_mark)
        
        #Now i ask the player what position they want to mark. Remember that intial list i made at the start?
        #I then edit that list as the player marks them. 
        #So if the player marks Position 1, the first position in that list will change i.e index 0

        def position_edit():
            choice = "F"
            while choice not in list(range(1,11,1)):
                choice= input("Please choose position to edit (1-10) ")
                choice= int(choice)
        
                if choice not in list(range(1,11,1)):
                    print ("Please choose position again! \n")
         
        
    
            return choice-1
    
        edited_position= position_edit()

        #Here i just save that initial list to a new list that i can edit. 
        #This is because I want the players to have the initial list everytime the game restarts
        new_positions =initial_positions


        #Now i mark the position in that list with the players choice
        new_positions[edited_position]= mark
        
        #Here i assign my variables to the positions as they are marked. This will keep updating as the while loop runs

        (a, b, c, d, e, f, g, h, i) = (new_positions)
        
        
        
        
        
                
        #Now i updated the list that while loop will run off when it starts again 
        initial_positions= new_positions
        
        #Now i need to check if someone has won so i will assign the rows, columns and diagonals of the table to variables
        
        row1_var= [a, b, c]
        row2_var= [d, e, f]
        row3_var= [g, h, i]
        col1_var= [a, d, g]
        col2_var= [b, e, h]
        col3_var= [c, f, i]
        diag1_var= [a, e, i]
        diag2_var= [c, e, g]
        
        #Then i make a function that checks if the rows, columns or diagonals are filled with the same letter
        
        def game_win(row_var): 
            if row_var== ["X", "X", "X"]:
                return "Y"
            elif row_var == ["O", "O", "O"]:
                return "N"
            else:
                return "N"
            
        
        game_win_row1= game_win(row1_var)
        game_win_row2= game_win(row2_var)
        game_win_row3= game_win(row3_var)
        game_win_col1= game_win(col1_var)
        game_win_col2= game_win(col2_var)
        game_win_col3= game_win(col3_var)
        game_win_diag1= game_win(diag1_var)
        game_win_diag2= game_win(diag2_var)
        
        #Now i show them the board 
        
        print(display_board_row1.format(a=a, b=b, c=c))
        print(display_board_row2.format(d=d, e=e, f=f))
        print(display_board_row3.format(g=g, h=h, i=i))
        
        #Now i check if someone has won
        
        if "Y" in [game_win_row1, game_win_row2, game_win_row3, game_win_col1, game_win_col2, game_win_col3, game_win_diag1, game_win_diag2]:
            game_win_check = "Y"
        elif "Y" not in [game_win_row1, game_win_row2, game_win_row3, game_win_col1, game_win_col2, game_win_col3, game_win_diag1, game_win_diag2]:
            game_win_check = "N"
        
         
        #Here i check on whose turn the game was won and then i tell them. Sweet victory! 
        
        if game_win_check == "Y" and play_count %2 == 0:
            print("Player 2 wins! \n")
        elif game_win_check == "Y" and play_count %2 != 0:
            print("Player 1 wins! \n")
            
        
    #Here if no one wins after all postions are filled, i tell them game is a draw
    if game_win_check == "N":
        print("Game is a draw! Rivals now? Who will eventually come out on top? Keep playing to find out! \n")
    
    #Here after someone has won or the table has been filled, I check if the players want to keep playing and update the choice
    
    def game_on(): 
        choice= False
        while choice not in ["Y", "N"]:
             choice=input("Would you like to keep playing? Y or N ")
        
             if choice not in ["Y", "N"]:
                 print ("Error! Please type Y or N to keep playing.")
        return choice
    
    game_on_choice= game_on()
    



        
        
    
#print(display_board_row1.format(a = "abx", b = "bbx", c = "abx"))
```
